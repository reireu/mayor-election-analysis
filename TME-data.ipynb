```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
df = pd.read_csv('reiwa02_TME.csv')
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['投票率_男', '投票率_女']], palette='Set3')
plt.title('男女別投票率の比較')
plt.xlabel('性別')
plt.ylabel('投票率 (%)')
plt.xticks(range(2), ['男性', '女性'])
plt.show()
```

```python
mean_rate = df['投票率_計'].mean()
print("投票率の平均値:", mean_rate)
```

```python
import matplotlib.pyplot as plt

# インデックスをリセット
df.reset_index(drop=True, inplace=True)

plt.figure(figsize=(20, 12))
plt.scatter(df['有権者数_計'], df['投票率_計'])

# 各点に投票所名を表示
for i, txt in enumerate(df['投票所名']):
    plt.annotate(txt, (df['有権者数_計'][i], df['投票率_計'][i]), fontsize=8)

# 投票率の平均値に水平線を引く
plt.axhline(y=56.224878048780496, color='red', linestyle='--', linewidth=1)

plt.xlabel('有権者数_計')
plt.ylabel('投票率_計')
plt.title('有権者数と投票率の関係')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

df['投票者数_男_女比'] = df['投票者数_男'] / df['投票者数_計']
df['投票者数_女_男比'] = df['投票者数_女'] / df['投票者数_計']

# 比率でソート
df_sorted = df.sort_values('投票者数_男_女比', ascending=False)

# 棒グラフを描画
plt.figure(figsize=(12, 10))
bar_width = 0.4
index = range(len(df_sorted))

# 男の比率
plt.barh(index, df_sorted['投票者数_男_女比'], height=bar_width, color='skyblue', label='男')

# 女の比率
plt.barh(index, df_sorted['投票者数_女_男比'], height=bar_width, color='pink', left=df_sorted['投票者数_男_女比'], label='女')

plt.yticks(index, df_sorted['投票所名'])
plt.xlabel('投票者数 男/女')
plt.title('投票者数の男女比')
plt.xlim(0, 1)

# 0.5のところに縦線を引く
plt.axvline(x=0.5, color='red', linestyle='--', linewidth=1)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
# データを整形
df_melted = df.melt(id_vars=['投票所名'], value_vars=['投票者数_男', '投票者数_女'], 
                     var_name='性別', value_name='投票者数')

# 棒グラフを描画
plt.figure(figsize=(14, 8))
sns.barplot(x='投票所名', y='投票者数', hue='性別', data=df_melted, palette='coolwarm')
plt.xticks(rotation=90)
plt.ylabel('投票者数')
plt.title('投票所ごとの男女別投票者数')
plt.legend(title='性別')
plt.tight_layout()
plt.show()
```

```python
# データを整形
df_melted = df.melt(id_vars=['投票所名'], value_vars=['投票率_男', '投票率_女'], 
                     var_name='性別', value_name='投票率')

# 棒グラフを描画
plt.figure(figsize=(14, 8))
sns.barplot(x='投票所名', y='投票率', hue='性別', data=df_melted, palette='coolwarm')
plt.xticks(rotation=90)
plt.ylabel('投票率')
plt.title('投票所ごとの男女別投票率')
plt.legend(title='性別')
plt.tight_layout()
plt.show()
```

```python
max_voting_rate = df['投票率_男'].max()
print("投票率_男の最大値:", max_voting_rate)

max_voting_rate = df['投票率_女'].max()
print("投票率_女の最大値:", max_voting_rate)
```

```python
# 相関行列の作成
correlation_matrix = df[['投票者数_男', '投票者数_女', '有権者数_計', '投票率_計']].corr()

# 相関行列のヒートマップ表示
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('相関行列')
plt.show()
```
